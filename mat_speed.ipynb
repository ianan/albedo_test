```python
import numpy as np, timeit

n, m = 2000, 400
spec = np.random.rand(n)
widths = np.random.rand(n)
matrix = np.random.rand(n, m)

def a():
    return np.einsum("i,ij,i->j", spec, matrix, widths, optimize=True)

def b():
    return np.matmul(spec, widths[:, None] * matrix)

# speed without the scaling
def c():
    return np.matmul(spec, matrix)

for fn in (a, b, c):
    t = timeit.timeit(fn, number=200)
    print(fn.__name__, f'{t:.4f}s')
```
